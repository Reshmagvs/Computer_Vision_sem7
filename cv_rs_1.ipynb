```python
import cv2
from google.colab.patches import cv2_imshow


image = cv2.imread('sample_data/cv-apple.jpg')

if image is None:
    print("Error: Could not find the image. Please make sure cv-apple.jpg is inside the sample_data folder.")
else:
    print("Original Image:")
    cv2_imshow(image)


    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    print("Grayscale Image:")
    cv2_imshow(gray_image)


    resized_image = cv2.resize(image, (200, 200))
    print("Resized Image:")
    cv2_imshow(resized_image)


    blurred_image = cv2.GaussianBlur(image, (15, 15), 0)
    print("Blurred Image:")
    cv2_imshow(blurred_image)


    edges = cv2.Canny(gray_image, 100, 200)
    print("Canny Edges:")
    cv2_imshow(edges)


    cv2.rectangle(image, (50, 50), (300, 300), (255, 0, 0), 2)
    cv2.line(image, (60, 60), (300, 300), (0, 0, 255), 2)
    print("Annotated Image Shapes:")
    cv2_imshow(image)
```
